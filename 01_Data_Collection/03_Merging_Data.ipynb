```python
import pandas as pd
```

```python
sap_df = pd.read_json(r"C:\Users\mehat\OneDrive - SRH\Subject\NLP\data\sap_news.json")
gnews_df = pd.read_json(r"C:\Users\mehat\OneDrive - SRH\Subject\NLP\data\gnews_articles.json")
```

```python
sap_df.rename(
    columns={"link": "url"},
    inplace=True
)
```

```python
sap_df["description"] = ""
```

```python
cols = [
    "title",
    "description",
    "content",
    "published",
    "source",
    "url"
]

sap_df = sap_df[cols]
gnews_df = gnews_df[cols]
```

```python
all_docs = pd.concat(
    [sap_df, gnews_df],
    ignore_index=True
)
```

```python
all_docs.shape
```

```python
all_docs['text'] = (all_docs['title'] + " " + all_docs["description"] + " " + all_docs["content"])
```

```python
print(all_docs.shape)

all_docs["text_length"] = all_docs["text"].str.len()

all_docs["text_length"].describe()
```

```python
all_docs.to_json(
    r"C:\Users\mehat\OneDrive - SRH\Subject\NLP\data\Merged_data.json",
    orient="records",
    indent=4
)
```
