```python
import feedparser
import pandas as pd
```

```python
feed = feedparser.parse(
    "https://news.sap.com/feed/"
)

len(feed.entries)
```

```python
sap_docs = []

for entry in feed.entries:
    
    sap_docs.append({
        "title": entry.title,
        "link": entry.link,
        "published": entry.published,
        "source": "SAP News"
    })

sap_df = pd.DataFrame(sap_docs)

sap_df.head()
```

```python
from newspaper import Article

contents = []

for url in sap_df["link"]:

    try:

        article = Article(url)

        article.download()
        article.parse()

        contents.append(article.text)

    except:

        contents.append("")
```

```python
sap_df["content"] = contents
```

```python
sap_df.to_json(
    r"C:\Users\mehat\OneDrive - SRH\Subject\NLP\data\sap_news.json",
    orient="records",
    indent=2
)
```
