### Merging all the collected resources (SAP News + GNews + Hacker News)

```python
import pandas as pd
```

```python
sap_df = pd.read_json("../notebook/data/sap_news.json")
gnews_df = pd.read_json("../notebook/data/gnews_articles.json")
hn_df = pd.read_json("../notebook/data/hackernews_posts.json")
```

```python
cols = [
    "title",
    "description",
    "content",
    "published",
    "source",
    "url"
]

sap_df = sap_df[cols]
gnews_df = gnews_df[cols]
hn_df = hn_df[cols]
```

```python
all_docs = pd.concat(
    [sap_df, gnews_df, hn_df],
    ignore_index=True
)

all_docs.shape
```

```python
# fillna first so one missing field doesn't blank out the whole text column
all_docs[["title", "description", "content"]] = all_docs[["title", "description", "content"]].fillna("")

all_docs['text'] = (all_docs['title'] + " " + all_docs["description"] + " " + all_docs["content"])
```

```python
all_docs["text_length"] = all_docs["text"].str.len()

all_docs["text_length"].describe()
```

```python
# quick check against the task 1 minimum requirement
print("total documents:", len(all_docs))
print("sources:", all_docs["source"].nunique(), "->", list(all_docs["source"].unique()))
```

```python
all_docs.to_json(
    "../notebook/data/Merged_data.json",
    orient="records",
    indent=4
)
```
